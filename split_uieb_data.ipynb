```python
import os
import shutil

# 原始影像來源資料夾
src_folder = "D:/GitHub/MobileIE/datasets/uieb/gt"

# 目標資料夾
train_folder = "D:/GitHub/MobileIE/datasets/uieb/train/gt"
test_folder = "D:/GitHub/MobileIE/datasets/uieb/test/gt"

# 如果資料夾不存在就建立
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# 遍歷來源資料夾內所有檔案
for filename in os.listdir(src_folder):
    src_path = os.path.join(src_folder, filename)

    # 如果不是檔案就跳過（例如資料夾）
    if not os.path.isfile(src_path):
        continue

    # 判斷是否含有 "img"
    if "img" in filename.lower():
        dst_path = os.path.join(train_folder, filename)
    else:
        dst_path = os.path.join(test_folder, filename)

    shutil.copy(src_path, dst_path)
    print(f"已複製：{filename} → {dst_path}")

print("完成分類！")
```
